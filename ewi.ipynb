```python
import pandas as pd
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
```

```python
bova=yf.download('^BVSP',
                 start='2014-01-01',
                 end='2021-12-31',
                 progress=False,
                 )
bova
```

```python
bova['Norm Close'] = bova['Close'] / bova['Close'][0]
bova
```

```python
for c in bova.columns:
    if c != 'Date':
        bova[c] = bova[c]/bova[c][0]
bova[c]
```

```python
bova.to_csv('bova.csv')
norm=pd.read_csv('bova.csv')
norm
```

```python
fig=px.line(norm,x='Date',y='Norm Close')
fig.show()
```
